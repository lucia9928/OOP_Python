Ejercicio 6. Constructor con parámetros
Crea una clase Punto con atributos x e y. Añade un método mostrar() que imprima las
coordenadas.

Ejemplo esperado:

p1 = Punto(3, 7)

p1.mostrar()

 Coordenadas: (3, 7)


```python
class Punto:
  def __init__(self, x, y):
    self.x=x
    self.y=y
  def mostrar(self):
    print("Coordenadas: ", self.x,",", self.y)
punto=Punto(5,8)
punto.mostrar()
```

Ejercicio 7. Modificación de atributos

Crea una clase Coche con un atributo color. Incluye un método

cambiar_color(nuevo_color) que modifique su valor.

Ejemplo:

c1 = Coche("rojo")

c1.cambiar_color("azul")

print(c1.color) # azul

```python
class Coche:
  def __init__(self, color):
    self.color= color
  def cambiar_color(self, nuevoColor):
    self.color=nuevoColor
coche=Coche("Azul")
print(coche.color)
coche.cambiar_color("Rojo")
print(coche.color)
```
